```python
def convert_duration_to_minutes(duration_str):
    total_minutes = 0
    if pd.isna(duration_str): # Agar NaN value hai to handle karo
        return np.nan # Ya koi aur default value jaise 0

    parts = duration_str.split(' ') # '2h', '50m' mein split karo

    for part in parts:
        if 'h' in part:
            hours = int(part.replace('h', ''))
            total_minutes += hours * 60
        elif 'm' in part:
            minutes = int(part.replace('m', ''))
            total_minutes += minutes
            
    return total_minutes

final_df['Duration_in_minutes'] = final_df['Duration'].apply(convert_duration_to_minutes)
```
